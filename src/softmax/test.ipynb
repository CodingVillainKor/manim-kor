```python
import torch
```

```python
def softmax(t, robust=False):
    if robust:
        t = t - torch.max(t)
    exps = torch.exp(t)
    sum_of_exps = torch.sum(exps)
    return exps / sum_of_exps

x1 = torch.tensor([0, 1, 2], dtype=torch.float16)
x2 = torch.tensor([10, 11, 12], dtype=torch.float16)
```

```python
result1 = softmax(x1, robust=True)
result2 = softmax(x2, robust=True)

print("Softmax of x1:", result1)
print("Softmax of x2:", result2)
```
